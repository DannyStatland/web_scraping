```python
import pandas as pd
from pandas_datareader import data as wb

stocks = []
f = open("sectors.txt","r")
for line in f:
    stocks.append(line.strip())
f.close()

wb.DataReader(stocks,"yahoo", start="2020-1-1",end="2021-11-05")["Adj Close"].to_csv("adjcloseSectors.csv")
```

```python
stocks = pd.read_csv('adjcloseSectors.csv')
stocks['Date'] = pd.to_datetime(stocks.Date)
stocks.set_index('Date',inplace=True)
stocks.pct_change().agg(['idxmin', 'min'])
```

```python
stocks_df = stocks.pct_change()
```

```python
stocks_df.tail()
```

```python
pd.to_numeric(stocks_df.iloc[1][:-2])
```

```python
stocks_df['best_sector'] = stocks_df.idxmax(axis=1)
stocks_df['worst_sector'] = stocks_df[stocks_df.columns.difference(['best_sector'])].idxmin(axis=1)
```
